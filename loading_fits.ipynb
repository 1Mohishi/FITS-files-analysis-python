# 01 Loading & Displaying FITS Files

This notebook covers the basics of loading a FITS file, reading its metadata, and displaying the raw image.

**Requirements:** astropy, numpy, matplotlib

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

fits_file = "your_image.fits"  # Change this to your file path

with fits.open(fits_file) as hdul:
    data   = hdul[0].data    # Raw pixel data
    header = hdul[0].header  # Observation metadata

print("Shape:", data.shape)
print("Min:",   np.min(data))
print("Max:",   np.max(data))
```

## Reading Metadata

The FITS header contains all observation details (telescope, location, date, exposure time, and more).

```python
# Print all metadata
for key, value in header.items():
    if key:
        print(f"{key:15} = {value}")
```

## Displaying the Raw Image

The raw image appears very dark most pixels are close to zero. This is normal in astrophotography.

```python
# Transpose from (3, H, W) to (H, W, 3) for matplotlib
rgb = np.transpose(data, (1, 2, 0))

plt.figure(figsize=(10, 6))
plt.imshow(rgb, origin='lower')
plt.title('Raw Image')
plt.axis('off')
plt.show()
```
